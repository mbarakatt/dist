```python
%matplotlib inline
from  pylab import *
import random
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.stats.stats import pearsonr
from mpl_toolkits.mplot3d import Axes3D
import scipy
```

```python
nbpts=100
pos=[[random.uniform(-50.0,50.0) for i in range(nbpts)] for j in [0,1]]
```

```python
def dist(p1,p2):
    return math.hypot(p2[0]-p1[0],p2[1]-p1[1])
def mapdistance(dist):
    #return min(10,np.random.poisson(1.0/(4000*math.pow(dist,2)  + 0.000000000001)))
    #Based on two-dimensional step stone model http://www.genetics.org/content/49/4/561.full.pdf
    return min(10,np.random.poisson(1.0/(math.exp(0.1*dist+1)*sqrt(dist+1))))
    #return 1.0/(log(dist)*sqrt(dist) + 1)
#def matrixmin(mat1,mat2):
#    return (mat1+mat2)/2 - np.abs(mat1-mat2)/2
def pointintersection(pos,myline):
    #return DistancePointLine(pos[0],pos[1],myline.a, myline.a*myline.slope + myline.y_inter,myline.b, myline.b*myline.slope + myline.y_inter,myline.length)
    return np.linalg.solve([[myline.slope,-1],[-myline.slope,-1]],[-myline.y_inter,-(pos[1]+myline.slope*pos[0])])
```

```python
class line:
    def __init__(self,x1,y1,x2,y2):# y_inter,theta,a=0,length=2.7):
        self.p1=[x1,y1]
        self.p2=[x2,y2]
        #print self.p1, self.p2
        if self.p1[0] == self.p2[0]:
                self.p2[0] += 0.0001
        self.length=dist(self.p1,self.p2)
        self.theta=math.atan((self.p2[0]-self.p1[0])/(self.p2[1]-self.p1[0]))#should be (-pi/2, pi/2)
        self.slope=math.tan(self.theta)
        #print self.slope
        self.y_inter=self.p1[1]-self.slope*self.p1[0]
    def __str__(self):
        return " ".join(map(str,[self.slope, self.theta, self.p1 , self.p2]))
def wrapper_line(params):
    return line(params[0],params[1],params[2],params[3])
```

```python
def DistancePointLine (pos, myline):
    pointinter=pointintersection(pos, myline)
    if pointinter[0] < min(myline.p1[0],myline.p2[0]) or pointinter[0] > max(myline.p1[0], myline.p2[0]): #or pointinter[1] < min(myline.p1[1],myline.p2[1]) or pointinter[1] > max(myline.p1[1],myline.p2[1]):
        return min(dist(pos,myline.p1),dist(pos,myline.p2))
    else:
        return dist(pointinter,pos)
    
    
    
```

```python
def compute_distances(pos,myline):
    dist_E=np.zeros([nbpts,nbpts])
    dist_T=np.zeros([nbpts,nbpts])
    dist_L=np.zeros([nbpts,nbpts])
    
    #print [pos[0][0], pos[1][0]] ,myline.a, myline.a*myline.slope + myline.y_inter,myline.b, myline.b*myline.slope + myline.y_inter,myline.length
    dist_line=[DistancePointLine([pos[0][i], pos[1][i]],myline) for i in range(nbpts)]
    #print dist_line[0]
    for i in range(nbpts):
        for j in range(i,nbpts):
            if i == j:
                continue
            dist_E[i][j]=dist([pos[0][i],pos[1][i]],[pos[0][j],pos[1][j]])
            dist_T[i][j]=dist_line[i] + dist_line[j]
            dist_L[i][j]=min(dist_E[i][j],dist_T[i][j])
    return dist_L

def compute_IBD(dist_L):
    IBD=np.zeros([nbpts,nbpts])
    plot(pos[0], pos[1],".")
    plot([myline.p1[0],myline.p2[0]],[myline.slope*myline.p1[0] + myline.y_inter ,myline.slope*myline.p2[0]+myline.y_inter])
    for i in range(nbpts):
        for j in range(i,nbpts):
            if i == j :
                continue
            IBD[i][j]=mapdistance(dist_L[i][j])
            if IBD[i][j]>0:
                print IBD[i][j]
                plot([pos[0][i],pos[0][j]],[pos[1][i],pos[1][j]], color='b',linewidth=math.log(IBD[i][j])+1)
    return IBD

def compute_corr(dist_L,IBD):
    a = np.ndarray.flatten(IBD)
    b = np.ndarray.flatten(dist_L)
    return pearsonr(a,b)[0]
def wrapper_compute_corr(params):
    y_inter=params[0]
    theta=params[1]
    a=params[2]
    length=params[3]
    #if y_inter > 1 or y_inter < -1 or slope < -1 or slope > 1:
        #return 100
    return compute_corr(compute_distances(pos,line(params[0],params[1],params[2],params[3])),IBD)
```
